```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import os
plt.style.use('ggplot')
plt.rcParams["figure.dpi"] = 150
```

```python
cars_df = pd.read_csv("USA_cars_datasets.csv", index_col=0)
cars_df.head()
```

```python
cars_df.describe()
```

```python
cars_df.info()
```

```python
cars_df.isnull().sum()
```

```python
cars_df = cars_df.drop(['vin', 'lot'], axis=1)
```

```python
cars_df.head()
```

```python
plt.bar(cars_df.brand.value_counts().index, cars_df.brand.value_counts(), width=0.5)
plt.xlabel("Brands")
plt.xticks(rotation=90)
plt.ylabel("Number of cars")
plt.show()
```

```python
plt.bar(cars_df.model.value_counts().index, cars_df.model.value_counts(), width=0.5)
plt.xlabel("Model")
plt.xticks(rotation=90, fontsize=5)
plt.ylabel("Number of cars")
plt.show()
```

```python
cars_df.model.value_counts()
```

```python
plt.bar(cars_df.state.value_counts().index, cars_df.state.value_counts(), width=0.5)
plt.xlabel("State")
plt.xticks(rotation=90, fontsize=10)
plt.ylabel("Number of cars")
plt.figure(figsize=(30, 20))
plt.show()
```

```python
plt.bar(cars_df.country.value_counts().index, cars_df.country.value_counts(), width=0.5)
plt.xlabel("Country")
plt.xticks(rotation=90)
plt.ylabel("Number of cars")
plt.show()
```

```python
cars_df.color.value_counts()
```

```python
plt.bar(cars_df.color.value_counts().index, cars_df.color.value_counts(), width=0.5)
plt.xlabel("Color of Car")
plt.xticks(rotation=90, fontsize=8)
plt.ylabel("Number of cars")
plt.show()
```
